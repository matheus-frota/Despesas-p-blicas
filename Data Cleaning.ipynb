<h1>Fase 1:Data Cleaning</h1>

```python
#Importando Bibliotecas
import pandas as pd
import numpy as np
from unicodedata import normalize
```

```python
# Unindo os bancos de dados.
def dataframes():
    
    # Criando dataframe vazio
    df_concat = pd.DataFrame()
    
    # Loop utilizado para percorrer todos os arquivos csv's.
    for i in range(1,32):
        
        #Condição utilizada para saltar o dia 6 que não teve despesa.
        if i < 10 and i != 6:
            
            # Variável que armazena o nome do arquivo csv.
            path = '2019010{}_Despesas_Pagamento.csv'.format(str(i))
            df = pd.read_csv(path,encoding = 'iso-8859-1',sep = ';')
            
            # Condição que só pega os arquivos com despesas, as outras são descartadas.
            if df.shape[0] > 0:
                
                # Concatenando os dataframes um a um.
                df_concat = pd.concat([df_concat,df])
            #print('Tamanho dos dados do dia {}: {}'.format(i,df.shape))
            #print('Tamanho concatenação: {}'.format(df_concat.shape))
            
        # Mesma ideia nessa condição, mas agora excluindo os dias 20,26 e 27.
        elif i >= 10 and i != 20 and i != 26 and i != 27:
            path = '201901{}_Despesas_Pagamento.csv'.format(str(i))
            df = pd.read_csv(path, encoding = 'iso-8859-1',sep = ';')
            if df.shape[0] > 0:
                df_concat = pd.concat([df_concat,df])
            #print('Tamanho dos dados do dia {}: {}'.format(i,df.shape))
            #print('Tamanho concatenação: {}'.format(df_concat.shape))
    #print('União {}'.format(df_concat.shape))
    
    # Retornando o dataframe concatenado
    return df_concat

df = dataframes()
```

<h3>Problemas econtrados:</h3>
    
* ParserError: ',' expected after '"': Resolvido alterando o parâmetro do delimitador do arquivo para ';'.
* UnicodeDecodeError: 'utf-8' codec can't decode byte 0xf3 in position 1: invalid continuation byte: Resolvido mudando a codificação utf-8 para iso-8859-1 que é a codificação dos caracteres latinos.

```python
# Alterando os nomes das colunas para facilitar sua manipulação

# Criando dicionario comprehensions para remover os acentos e escrever apenas a primeira letra como maiúscula.
# Na forma de dicionário para facilitar quando passar para o método rename -> {'Nome antigo':'Nome modificado'}
nome_colunas_modif = {val : normalize('NFKD', val).encode('ASCII','ignore').decode('ASCII').replace(' ','_').capitalize() for index,val in enumerate(df.columns)}

# Renomeando todas as colunas do dataframe
df = df.rename(index=str,columns = nome_colunas_modif)
```

```python
# Visualizando modificações
df.head()
```

```python
# Trocando o tipo das colunas de object para float
col_numerica = ['Valor_original_do_pagamento','Valor_do_pagamento_convertido_pra_r$','Valor_utilizado_na_conversao']

# Fazendo essa troca nas 3 colunas
for col in col_numerica:
    df[col] = df[col].str.replace(',','.').astype('float')

# Verificando se a troca funcionou
df.info()
```

```python
# Agora irei remover todas as colunas com o nome codigo embutido no nome.
df = df.drop([x for x in df.columns if x.count('Codigo')], axis = 1)
```

```python
# Após limpar as colunas irei substituir todos os valores ausentes da coluna 'Processo'.
df = df.fillna({'Processo':'Sem processo'})
```
